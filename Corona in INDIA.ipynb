```python
import pandas as pd
dd = pd.read_html('https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_India')
state_df = dd[5].iloc[:-4]
state_df.fillna(0, inplace=True)
state_df = state_df[state_df.columns[ 0 : 25 ]]
state_df.columns = ['Date', 'AP', 'CH', 'CT', 'DL', 'GJ', 'HR', 'HP', 'JK', 'KA', 'KL', 'LA', 'MP', 'MH', 'OR', 'PY', 'PB', 'RJ', 'TN', 'TG', 'UT', 'UP', 'WB', 'New', 'ALL']
state_df = state_df.drop(['New'], axis = 1) 
state_df[['Date', 'AP', 'ALL']] 
```
